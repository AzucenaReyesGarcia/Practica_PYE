### Ejercicio 2:

a) Se sabe que las calificaciones o puntajes de evaluaciones siguen una distribuci√≥n normal.
De acuerdo con los datos proporcionados en el fragmento anterior, ¬øcu√°l es la desviaci√≥n est√°ndar mexicana?

b) ¬øCu√°les son los puntajes en los que se encuentra el 90% central de los estudiantes mexicanos?


### Para resolver este ejercicio:

a) Calcular la desviaci√≥n est√°ndar mexicana:

Sabemos que:

- La media (ùúá) mexicana es 405.

- Para que un puntaje de 650 est√© en el 10% superior de la distribuci√≥n 
(ùëÉ(ùëã>650)=0.10), necesitamos calcular la desviaci√≥n est√°ndar (ùúé).

La f√≥rmula de estandarizaci√≥n de la distribuci√≥n normal es:

$$Z = \frac{X - \mu}{\sigma}$$

Usamos la funci√≥n inversa de la distribuci√≥n normal est√°ndar para encontrar ùëç, y luego resolvemos para ùúé.

---

b) Calcular los puntajes del 90% central:

- El 90% central corresponde al rango de puntajes entre el percentil 5 y el percentil 95. Esto significa que:

$$P(X \leq X_{\text{inferior}}) = 0.05 \quad \text{y} \quad P(X \leq X_{\text{superior}}) = 0.95$$

Calculamos los l√≠mites usando la desviaci√≥n est√°ndar obtenida.

```python
from scipy.stats import norm

# Datos
media_mexicana = 405  # Media mexicana
puntaje_superior = 650  # Puntaje del 10% superior
prob_superior = 0.10  # Probabilidad del 10% superior

z_superior = norm.ppf(1 - prob_superior)

desviacion_estandar = (puntaje_superior - media_mexicana) / z_superior

print(f"a) Desviaci√≥n est√°ndar mexicana: {desviacion_estandar:.2f}")

z_inferior = norm.ppf(0.05)
z_superior = norm.ppf(0.95)

puntaje_inferior = media_mexicana + z_inferior * desviacion_estandar
puntaje_superior = media_mexicana + z_superior * desviacion_estandar
print(f"b) Rango del 90% central de estudiantes mexicanos: {puntaje_inferior:.2f} a {puntaje_superior:.2f}")
```

### Ejercicio 3:
Como vimos en este bloque, el √°rea bajo la curva de una funci√≥n se traduce en la probabilidad entre dos valores. Calcula las integrales definidas siguientes:

a) $\int_{1}^{1.5} (2x + 1) \, dx$

b) $\int_{0}^{0.5} \frac{x^2}{3} \, dx$

c) $\int_{1}^{1.2} e^x \, dx$

d) $\int_{\frac{\pi}{6}}^{\frac{\pi}{4}} \sin x \, dx$




#### a) 
$\int_{1}^{1.5} (2x + 1) \, dx$}
\begin{align*}
\text{Antiderivada: } & \int (2x + 1) \, dx = x^2 + x + C \\
\text{Evaluamos: } & F(1.5) = (1.5)^2 + (1.5) = 2.25 + 1.5 = 3.75 \\
                   & F(1) = (1)^2 + (1) = 1 + 1 = 2 \\
\text{Integral: } & \int_{1}^{1.5} (2x + 1) \, dx = F(1.5) - F(1) = 3.75 - 2 = 1.75
\end{align*}

#### b) 
$\int_{0}^{0.5} \frac{x^2}{3} \, dx$}
\begin{align*}
\text{Antiderivada: } & \int \frac{x^2}{3} \, dx = \frac{x^3}{9} + C \\
\text{Evaluamos: } & F(0.5) = \frac{(0.5)^3}{9} = \frac{0.125}{9} = \frac{1}{72} \\
                   & F(0) = 0 \\
\text{Integral: } & \int_{0}^{0.5} \frac{x^2}{3} \, dx = F(0.5) - F(0) = \frac{1}{72} - 0 = \frac{1}{72}
\end{align*}

#### c) 
$\int_{1}^{1.2} e^x \, dx$}
\begin{align*}
\text{Antiderivada: } & \int e^x \, dx = e^x + C \\
\text{Evaluamos: } & F(1.2) = e^{1.2} \\
                   & F(1) = e^1 \\
\text{Integral: } & \int_{1}^{1.2} e^x \, dx = F(1.2) - F(1) = e^{1.2} - e
\end{align*}

#### d) 
$\int_{\frac{\pi}{6}}^{\frac{\pi}{4}} \sin x \, dx$}
\begin{align*}
\text{Antiderivada: } & \int \sin x \, dx = -\cos x + C \\
\text{Evaluamos: } & F\left(\frac{\pi}{4}\right) = -\cos\left(\frac{\pi}{4}\right) = -\frac{\sqrt{2}}{2} \\
                   & F\left(\frac{\pi}{6}\right) = -\cos\left(\frac{\pi}{6}\right) = -\frac{\sqrt{3}}{2} \\
\text{Integral: } & \int_{\frac{\pi}{6}}^{\frac{\pi}{4}} \sin x \, dx = F\left(\frac{\pi}{4}\right) - F\left(\frac{\pi}{6}\right) = -\frac{\sqrt{2}}{2} + \frac{\sqrt{3}}{2}
\end{align*}



```python
import sympy as sp

x = sp.Symbol('x')

integral_a = sp.integrate(2 * x + 1, (x, 1, 1.5))

integral_b = sp.integrate(x**2 / 3, (x, 0, 0.5))

integral_c = sp.integrate(sp.exp(x), (x, 1, 1.2))

integral_d = sp.integrate(sp.sin(x), (x, sp.pi / 6, sp.pi / 4))

print(f"a) Integral de (2x + 1) dx desde 1 a 1.5: {integral_a}")
print(f"b) Integral de (x^2 / 3) dx desde 0 a 0.5: {integral_b}")
print(f"c) Integral de e^x dx desde 1 a 1.2: {integral_c.evalf()}")
print(f"d) Integral de sen(x) dx desde œÄ/6 a œÄ/4: {integral_d.evalf()}")
```

### Ejercicio 4:

c) P(‚àí0.1 ‚â§ Z ‚â§ 0.9)

d) P(‚àí3.66 ‚â§ Z ‚â§ ‚àí2.15)




Sea \( Z \) una variable aleatoria que tiene una distribuci√≥n normal est√°ndar; encuentra:

####  a)   P(Z ‚â§ 1.54)

\begin{align*}
\text{Utilizando la tabla de la distribuci√≥n normal est√°ndar, encontramos:} & \\
P(Z \leq 1.54) & \approx 0.9382
\end{align*}


####  b) P( Z > 1.54)

\begin{align*}
P(Z > 1.54) & = 1 - P(Z \leq 1.54) \\
             & \approx 1 - 0.9382 = 0.0618
\end{align*}


####  c) P(‚àí0.1 ‚â§ Z ‚â§ 0.9)

\begin{align*}
P(-0.1 \leq Z \leq 0.9) & = P(Z \leq 0.9) - P(Z \leq -0.1) \\
\text{Buscamos estos valores:} & \\
P(Z \leq 0.9) & \approx 0.8159 \\
P(Z \leq -0.1) & \approx 0.4602 \\
P(-0.1 \leq Z \leq 0.9) & \approx 0.8159 - 0.4602 = 0.3557
\end{align*}


####  d) P(‚àí3.66 ‚â§ Z ‚â§ ‚àí2.15)

\begin{align*}
P(-3.66 \leq Z \leq -2.15) & = P(Z \leq -2.15) - P(Z \leq -3.66) \\
\text{Buscamos estos valores:} & \\
P(Z \leq -2.15) & \approx 0.0158 \\
P(Z \leq -3.66) & \approx 0.0001 \\
P(-3.66 \leq Z \leq -2.15) & \approx 0.0158 - 0.0001 = 0.0157
\end{align*}




```python
from scipy.stats import norm

p_a = norm.cdf(1.54)

p_b = 1 - norm.cdf(1.54)

p_c = norm.cdf(0.9) - norm.cdf(-0.1)

p_d = norm.cdf(-2.15) - norm.cdf(-3.66)

print(f"a) P(Z ‚â§ 1.54): {p_a:.4f}")
print(f"b) P(Z > 1.54): {p_b:.4f}")
print(f"c) P(‚àí0.1 ‚â§ Z ‚â§ 0.9): {p_c:.4f}")
print(f"d) P(‚àí3.66 ‚â§ Z ‚â§ ‚àí2.15): {p_d:.4f}")
```

### Ejercicio 6:
Sea X una variable aleatoria que se distribuye de manera normal con Œº = 4 y œÉ = 1.5. Calcula:

a) P(X ‚â§ 5)

b) P(X ‚â• 4.5)

c) P(0 ‚â§ X ‚â§ 2)



Sea \( X \) una variable aleatoria que se distribuye de manera normal con \( \mu = 4 \) y \( \sigma = 1.5 \). Calcula:

#### a) P(X ‚â§ 5)

\begin{align*}
Z & = \frac{5 - 4}{1.5} \approx 0.67 \\
\text{Buscando en la tabla de la distribuci√≥n normal est√°ndar, encontramos:} & \\
P(Z \leq 0.67) & \approx 0.7486
\end{align*}

#### b) P(X ‚â• 4.5)

\begin{align*}
Z & = \frac{4.5 - 4}{1.5} \approx 0.33 \\
P(X \geq 4.5) & = 1 - P(Z \leq 0.33) \approx 1 - 0.6293 \approx 0.3707
\end{align*}

#### c) P(0 ‚â§ X ‚â§ 2)

\begin{align*}
Z_0 & = \frac{0 - 4}{1.5} \approx -2.67 \\
Z_2 & = \frac{2 - 4}{1.5} \approx -1.33 \\
P(0 \leq X \leq 2) & = P(Z \leq -1.33) - P(Z \leq -2.67) \\
P(Z \leq -1.33) & \approx 0.0918 \\
P(Z \leq -2.67) & \approx 0.0038 \\
P(0 \leq X \leq 2) & \approx 0.0918 - 0.0038 \approx 0.0880
\end{align*}

```python
from scipy.stats import norm

mu = 4 
sigma = 1.5

p_a = norm.cdf(5, loc=mu, scale=sigma)

p_b = 1 - norm.cdf(4.5, loc=mu, scale=sigma)

p_c = norm.cdf(2, loc=mu, scale=sigma) - norm.cdf(0, loc=mu, scale=sigma)

# Imprimir resultados
print(f"a) P(X ‚â§ 5): {p_a:.4f}")
print(f"b) P(X ‚â• 4.5): {p_b:.4f}")
print(f"c) P(0 ‚â§ X ‚â§ 2): {p_c:.4f}")
```

### Ejercicio 7:
De acuerdo con la Organizaci√≥n de Naciones Unidas (onu), en Latinoam√©rica se tiene una esperanza de vida al nacer de 73.4 a√±os y una desviaci√≥n est√°ndar de 4.3714 a√±os. 

Si la esperanza de vida de una persona sigue una distribuci√≥n normal, calcula:

a) La probabilidad de que una persona sobrepase los 80 a√±os de vida.

b) La probabilidad de que una persona al nacer tenga una esperanza de vida entre 70 y 75 a√±os.

c) ¬øA partir de qu√© valor se tiene el 10% m√°s alto de la esperanza de vida al nacer?



#### a) La probabilidad de que una persona sobrepase los 80 a√±os de vida.

\begin{align*}
Z & = \frac{80 - 73.4}{4.3714} \approx 1.5 \\
P(X > 80) & = 1 - P(Z \leq 1.5) \\
& \approx 1 - 0.9332 \approx 0.0668
\end{align*}

#### b) La probabilidad de que una persona al nacer tenga una esperanza de vida entre 70 y 75 a√±os.

\begin{align*}
Z_{70} & = \frac{70 - 73.4}{4.3714} \approx -0.78 \\
Z_{75} & = \frac{75 - 73.4}{4.3714} \approx 0.36 \\
P(70 < X < 75) & = P(Z \leq 0.36) - P(Z \leq -0.78) \\
& \approx 0.6406 - 0.2177 \approx 0.4229
\end{align*}

#### c) ¬øA partir de qu√© valor se tiene el 10\% m√°s alto de la esperanza de vida al nacer?

Para encontrar el percentil 90:

\begin{align*}
P(Z \leq z) & = 0.90 \\
z & \approx 1.2816 \\
X & = \mu + z \cdot \sigma \\
& = 73.4 + 1.2816 \cdot 4.3714 \approx 77.47
\end{align*}






### Para resolver este ejercicio:

Datos del problema:

- Media (ùúá) = 73.4 a√±os.
- Desviaci√≥n est√°ndar (ùúé) = 4.3714 a√±os.

F√≥rmulas clave:

- La probabilidad acumulada se calcula como ùëÉ(ùëã‚â§ùë•) usando la funci√≥n acumulada de la normal.

- Usamos la f√≥rmula de estandarizaci√≥n

$$Z = \frac{X - \mu}{\sigma}$$


```python
from scipy.stats import norm

mu = 73.4  # esperanza de vida (media)
sigma = 4.3714  # desviaci√≥n est√°ndar

p_a = 1 - norm.cdf(80, loc=mu, scale=sigma)

p_b = norm.cdf(75, loc=mu, scale=sigma) - norm.cdf(70, loc=mu, scale=sigma)

percentile_90 = norm.ppf(0.9, loc=mu, scale=sigma)

# Imprimir resultados
print(f"a) P(X > 80): {p_a:.4f}")
print(f"b) P(70 ‚â§ X ‚â§ 75): {p_b:.4f}")
print(f"c) Valor para el 10% m√°s alto de la esperanza de vida: {percentile_90:.2f} a√±os")
```

### Ejercicio 8:

Entre mayo y junio de 2011, la cotizaci√≥n del d√≥lar a la venta tuvo una media de $16.9205 y una desviaci√≥n est√°ndar de $0.2695. Si la cotizaci√≥n del d√≥lar sigue una distribuci√≥n normal, determina:

a) La probabilidad de que la venta de un d√≥lar sobrepase los $18.

b) La probabilidad de que la cotizaci√≥n del d√≥lar a la venta sea menor a los $15.



#### a) La probabilidad de que la venta de un d√≥lar sobrepase los $18.

\begin{align*}
Z & = \frac{18 - 16.9205}{0.2695} \approx 4.008 \\
P(X > 18) & = 1 - P(Z \leq 4.008) \\
& \approx 1 - 0.9999 \approx 0.0001
\end{align*}

#### b) La probabilidad de que la cotizaci√≥n del d√≥lar a la venta sea menor a los $15.

\begin{align*}
Z & = \frac{15 - 16.9205}{0.2695} \approx -7.12 \\
P(X < 15) & = P(Z \leq -7.12) \\
& \approx 0.0000
\end{align*}



### Para resolver este ejercicio:

Datos del problema:

- Media (ùúá) = 16.9205.

- Desviaci√≥n est√°ndar (ùúé) = 0.2695.

F√≥rmulas clave:

- Para calcular ùëÉ(ùëã‚â§ùë•), usamos la funci√≥n acumulada de la distribuci√≥n normal: norm.cdf(x,¬†loc=Œº,¬†scale=œÉ).


- Para probabilidades de tipo ùëÉ(ùëã>ùë•), usamos la relaci√≥n complementaria: 
ùëÉ(ùëã>ùë•)=1‚àíùëÉ(ùëã‚â§ùë•).


```python
from scipy.stats import norm

mu = 16.9205  # Media
sigma = 0.2695  # Desviaci√≥n est√°ndar

p_a = 1 - norm.cdf(18, loc=mu, scale=sigma)

p_b = norm.cdf(15, loc=mu, scale=sigma)

print(f"a) P(X > 18): {p_a:.4f}")
print(f"b) P(X < 15): {p_b:.4f}")
```

- Para calcular un intervalo $$P(a \leq X \leq b)$$

# Ejercicio 13: Horas de Sue√±o

La falta de sue√±o reparador puede afectar nuestra capacidad para llevar a cabo las actividades diarias, ya que estamos demasiado cansados o tenemos problemas de concentraci√≥n. Todos los tipos de insomnio pueden llevar a somnolencia diurna, mala concentraci√≥n e incapacidad para sentirnos renovados y descansados en la ma√±ana.

## a) Pregunta a tus compa√±eros de clase cu√°ntas horas duermen al d√≠a.

- Este paso es emp√≠rico y depende de los datos que hayas recogido de tus compa√±eros de clase. Una vez que tengas esos datos, puedes calcular la **media** y la **desviaci√≥n est√°ndar**.

## b) Calcular la media y la desviaci√≥n est√°ndar poblacionales.

La **media poblacional** $\mu$ y la **desviaci√≥n est√°ndar poblacional** $\sigma$ se calculan con las siguientes f√≥rmulas:

### Media poblacional $\mu$:

$$
\mu = \frac{1}{N} \sum_{i=1}^{N} X_i
$$

Donde:
- \(N\) es el n√∫mero total de personas (el tama√±o de la muestra o poblaci√≥n).
- \(X_i\) es el n√∫mero de horas que duerme cada persona.

### Desviaci√≥n est√°ndar poblacional $\sigma$:

$$
\sigma = \sqrt{\frac{1}{N} \sum_{i=1}^{N} (X_i - \mu)^2}
$$

#### Ejemplo con datos:
Supongamos que recopilaste las siguientes horas de sue√±o de 5 compa√±eros de clase: 5, 7, 6, 8, 5.

1. **Calcular la media**:
   $$
   \mu = \frac{1}{5} (5 + 7 + 6 + 8 + 5) = \frac{31}{5} = 6.2
   $$

2. **Calcular la desviaci√≥n est√°ndar**:
   - Primero, calculamos las diferencias al cuadrado respecto a la media:
     $$
     (5 - 6.2)^2 = 1.44, \quad (7 - 6.2)^2 = 0.64, \quad (6 - 6.2)^2 = 0.04, \quad (8 - 6.2)^2 = 3.24, \quad (5 - 6.2)^2 = 1.44
     $$
   - Sumamos los valores al cuadrado:
     $$
     1.44 + 0.64 + 0.04 + 3.24 + 1.44 = 6.8
     $$
   - Ahora calculamos la desviaci√≥n est√°ndar:
     $$
     \sigma = \sqrt{\frac{6.8}{5}} = \sqrt{1.36} \approx 1.167
     $$

Por lo tanto, la media es \(6.2\) horas y la desviaci√≥n est√°ndar es aproximadamente \(1.167\) horas.

## c) ¬øQu√© porcentaje de tus compa√±eros duermen menos de seis horas?

Para calcular el porcentaje de compa√±eros que duermen menos de seis horas, utilizamos la **distribuci√≥n normal**. Para esto, necesitamos el valor de \(Z\) correspondiente a las horas de sue√±o \(X = 6\), usando la f√≥rmula de transformaci√≥n est√°ndar:

$$
Z = \frac{X - \mu}{\sigma}
$$

Sustituyendo los valores conocidos:

$$
Z = \frac{6 - 6.2}{1.167} = \frac{-0.2}{1.167} \approx -0.171
$$

- Usamos la tabla de la distribuci√≥n normal est√°ndar o una calculadora para encontrar el valor acumulado para \(Z = -0.171\). Este valor es aproximadamente \(P(Z \leq -0.171) \approx 0.432\).

Por lo tanto, el porcentaje de compa√±eros que duermen menos de seis horas es aproximadamente:

$$
\text{Porcentaje} \approx 43.2\%
$$

## d) ¬øDebajo de qu√© valor se encuentra el 10\% de los que menos horas duermen?

Para encontrar el valor debajo del cual se encuentra el 10\% de los compa√±eros que duermen menos horas, buscamos el valor de \(Z\) correspondiente al percentil 10. El valor de \(Z\) para el 10\% en la distribuci√≥n normal es aproximadamente **-1.28**.

Usamos la f√≥rmula de transformaci√≥n para despejar \(X\):

$$
Z = \frac{X - \mu}{\sigma}
$$

Sustituyendo \(Z = -1.28\), \($\mu$ = 6.2\) y \($\sigma$ = 1.167\):

$$
-1.28 = \frac{X - 6.2}{1.167}
$$

Despejamos \(X\):

$$
X = -1.28 \times 1.167 + 6.2 = -1.493 + 6.2 \approx 4.71
$$

Por lo tanto, el valor debajo del cual se encuentra el 10\% de los compa√±eros que duermen menos horas es aproximadamente:

$$
X \approx 4.71 \text{ horas}
$$

c) El **porcentaje** de compa√±eros que duermen menos de 6 horas es aproximadamente **43.2\%**.

d) El valor debajo del cual se encuentra el **10\%** de los compa√±eros que duermen menos horas es **4.71 horas**.


```python
import numpy as np
import scipy.stats as stats

# Datos
horas_sue√±o = [7, 8, 6, 5, 7, 9, 5, 6, 8, 7, 5, 6, 8, 6, 5]

# C√°lculos
media = np.mean(horas_sue√±o)
desviacion_estandar = np.std(horas_sue√±o, ddof=1)  # Usar ddof=1 para una muestra

# C√°lculo del porcentaje de compa√±eros que duermen menos de 6 horas
porcentaje_menos_6 = stats.norm.cdf(6, loc=media, scale=desviacion_estandar) * 100

# C√°lculo del valor del percentil 10
valor_10_percentil = stats.norm.ppf(0.10, loc=media, scale=desviacion_estandar)

# Imprimir resultados
print(f"Media de horas de sue√±o: {media:.2f} horas")
print(f"Desviaci√≥n est√°ndar de horas de sue√±o: {desviacion_estandar:.2f} horas")
print(f"Porcentaje de compa√±eros que duermen menos de 6 horas: {porcentaje_menos_6:.2f}%")
print(f"Valor debajo del cual se encuentra el 10% de los que menos duermen: {valor_10_percentil:.2f} horas")
```

Por lo tanto, la desviaci√≥n est√°ndar es aproximadamente 
$$ \sigma \approx 14.63 $$

## b) Calcular la probabilidad de que una persona tenga un puntaje por debajo de 70 (retardo mental).

- Usamos la f√≥rmula de transformaci√≥n para convertir el puntaje $$X = 70$$ a la variable est√°nda $$Z:$$


$$Z = \frac{X - \mu}{\sigma} = \frac{70 - 100}{14.63} = \frac{-30}{14.63} \approx -2.05$$


- Buscamos en la tabla de la distribuci√≥n normal est√°ndar el valor correspondiente a $$Z = -2.05$$  Este valor es aproximadamente $$P(Z \leq -2.05) = 0.02\$$

Por lo tanto, la probabilidad de que una persona tenga un puntaje por debajo de 70 es:


$$P(X < 70) \approx 0.02$$


Esto significa que el 2\% de la poblaci√≥n tiene un puntaje inferior a 70.

## c) ¬øQu√© proporci√≥n de la poblaci√≥n se le considera "genio"? Una persona con un puntaje de 165 o mayor es generalmente considerada genio.

- Usamos la f√≥rmula de transformaci√≥n para convertir el puntaje \(X = 165\) a la variable est√°ndar \(Z\):


$$Z = \frac{X - \mu}{\sigma} = \frac{165 - 100}{14.63} = \frac{65}{14.63} \approx 4.44$$


- Buscamos en la tabla de la distribuci√≥n normal est√°ndar el valor correspondiente a \(Z = 4.44\). Este valor es muy cercano a 1, lo que indica que la probabilidad acumulada es pr√°cticamente 1.
- La probabilidad de que una persona tenga un puntaje de 165 o mayor es:


$$P(X \geq 165) = 1 - P(Z \leq 4.44) = 1 - 0.99999 = 0.00001$$


Por lo tanto, la proporci√≥n de la poblaci√≥n considerada genio (puntaje \(\geq 165\)) es aproximadamente \(0.00001\%\).


### Pasos para resolver este ejercicio:


a) Desviaci√≥n est√°ndar:
Sabemos que el cociente intelectual sigue una distribuci√≥n normal con una media Œº=100, y que el 2% de las personas tienen un puntaje mayor a 130, lo que corresponde al 98% en la cola izquierda de la distribuci√≥n normal.

El puntaje de 130 est√° en el percentil 98. Para determinar la desviaci√≥n est√°ndar (ùúé), podemos usar la f√≥rmula de la distribuci√≥n normal est√°ndar:

$$Z = \frac{X - \mu}{\sigma}$$
‚Äã
Donde: Z es el valor cr√≠tico para el percentil 98 (usaremos una tabla de la distribuci√≥n normal est√°ndar o una funci√≥n de Python para encontrarlo), X=130 es el puntaje de cociente intelectual, Œº=100 es la media.

Parte b) Probabilidad de tener un puntaje por debajo de 70:
La probabilidad de que una persona tenga un puntaje por debajo de 70 se puede encontrar utilizando la f√≥rmula de la distribuci√≥n normal est√°ndar. Se necesita calcular el valor ùëç para X=70 y luego usarlo para obtener la probabilidad asociada.

Parte c) Proporci√≥n de la poblaci√≥n con un puntaje de 165 o mayor:
Para encontrar la proporci√≥n de personas con un puntaje de 165 o mayor, se calcula el valor 
Z correspondiente a X=165 y luego se determina la probabilidad asociada con ese valor.


```python
import scipy.stats as stats

# Funci√≥n para calcular la desviaci√≥n est√°ndar
def calcular_desviacion_estandar(media, puntaje_superdotado, percentil):
    # Calculamos el valor Z correspondiente al percentil
    Z = stats.norm.ppf(percentil, loc=media, scale=1)
    # Desviaci√≥n est√°ndar es (X - media) / Z
    desviacion_estandar = (puntaje_superdotado - media) / (Z - media)
    return desviacion_estandar

# Funci√≥n para calcular la probabilidad de puntaje por debajo de un valor
def probabilidad_por_bajo(puntaje, media, desviacion_estandar):
    return stats.norm.cdf(puntaje, loc=media, scale=desviacion_estandar)

# Funci√≥n para calcular la probabilidad de puntaje mayor o igual a un valor
def probabilidad_por_encima(puntaje, media, desviacion_estandar):
    return 1 - stats.norm.cdf(puntaje, loc=media, scale=desviacion_estandar)

# Par√°metros del problema
media = 100
puntaje_superdotado = 130
puntaje_bajo = 70
puntaje_genio = 165
percentil = 0.98  # El 98% de las personas est√°n por debajo de 130

# Parte a) Calcular la desviaci√≥n est√°ndar
desviacion_estandar = calcular_desviacion_estandar(media, puntaje_superdotado, percentil)

# Parte b) Calcular la probabilidad de puntaje por debajo de 70
probabilidad_bajo_70 = probabilidad_por_bajo(puntaje_bajo, media, desviacion_estandar)

# Parte c) Calcular la probabilidad de puntaje mayor o igual a 165
probabilidad_mayor_165 = probabilidad_por_encima(puntaje_genio, media, desviacion_estandar)

# Imprimir resultados
print(f"Desviaci√≥n est√°ndar: {desviacion_estandar:.2f}")
print(f"Probabilidad de puntaje por debajo de 70: {probabilidad_bajo_70:.4f}")
print(f"Proporci√≥n de personas consideradas 'genio' (puntaje >= 165): {probabilidad_mayor_165:.8f}")
```

a) La probabilidad de que una pel√≠cula dure menos de 90 minutos.

b) La probabilidad de que una pel√≠cula dure entre 95 y 125 minutos.

c) La probabilidad de que dure m√°s de 150 minutos.

### Pasos para resolver este ejercicio:
Datos del problema:
Media (ùúá) = 112.5111 minutos

Varianza ($ùúé^2$): 383.8054

Desviaci√≥n est√°ndar (ùúé): $\sqrt{\text{Varianza}} = \sqrt{383.8054}$

F√≥rmulas clave:

Utilizamos la distribuci√≥n normal est√°ndar para calcular probabilidades. La f√≥rmula de estandarizaci√≥n es:

$$Z = \frac{X - \mu}{\sigma}$$

Con Z, calculamos las probabilidades con las funciones acumuladas $1‚àínorm.cdf(Z)$, seg√∫n corresponda.

----

a) Probabilidad de que una pel√≠cula dure menos de 90 minutos:

$$P(X‚â§90)$$

- Convertimos X=90 a un valor ùëç:

$$Z = \frac{90 - \mu}{\sigma}$$
 
Calculamos P(X‚â§90) usando la funci√≥n acumulada $norm.cdf(Z)$.

----

b) Probabilidad de que una pel√≠cula dure entre 95 y 125 minutos:

$$P(95‚â§X‚â§125)=P(X‚â§125)‚àíP(X‚â§95)$$

- Convertimos ùëã=95 y ùëã=125 a sus valores correspondientes de ùëç:

$$Z_{95} = \frac{95 - \mu}{\sigma}, \quad Z_{125} = \frac{125 - \mu}{\sigma}$$

- Calculamos ambas probabilidades acumuladas y hacemos la resta:

$$P(95‚â§X‚â§125)=norm.cdf(Z_{125})‚àínorm.cdf(Z_{95})$$

---

c) Probabilidad de que una pel√≠cula dure m√°s de 150 minutos:

$$P(X‚â•150)=1‚àíP(X‚â§150)$$

Convertimos X=150 a Z:
‚Äã
$$Z = \frac{150 - \mu}{\sigma}$$
 
Calculamos $P(X‚â§150)$ con $norm.cdf(Z)$ y restamos de 1.

```python
import scipy.stats as stats
import numpy as np

media = 112.5111
varianza = 383.8054
desviacion_estandar = np.sqrt(varianza)

probabilidad_menos_90 = stats.norm.cdf(90, loc=media, scale=desviacion_estandar)

probabilidad_entre_95_y_125 = stats.norm.cdf(125, loc=media, scale=desviacion_estandar) - stats.norm.cdf(95, loc=media, scale=desviacion_estandar)

probabilidad_mas_150 = 1 - stats.norm.cdf(150, loc=media, scale=desviacion_estandar)

# Imprimir resultados
print(f"Probabilidad de que una pel√≠cula dure menos de 90 minutos: {probabilidad_menos_90:.4f}")
print(f"Probabilidad de que una pel√≠cula dure entre 95 y 125 minutos: {probabilidad_entre_95_y_125:.4f}")
print(f"Probabilidad de que una pel√≠cula dure m√°s de 150 minutos: {probabilidad_mas_150:.4f}")
```

### Ejercicio 16: 
Cualquier tipo de mascota requiere cuidados y tiempo, adem√°s de que implica gastos de alimentaci√≥n y salud, entre otros. Por ello, se debe ser consciente al tener una mascota; por ejemplo, un perro. De acuerdo con las estad√≠sticas de refugios para perros callejeros, ocho de cada 10 perros que se compran terminan en la calle. Esto representa no s√≥lo un problema de salud sino de falta de responsabilidad por parte del due√±o. Si en el mes siguiente se venden 50 perros, determina, utilizando la aproximaci√≥n a la normal:

a) ¬øCu√°l es la probabilidad de que menos de 30 perros terminen en la calle?

b) ¬øCu√°l es la probabilidad de que entre 30 y 45 perros terminen en la calle?

### Pasos para resolver este ejercicio:

Datos del problema:

$n=50$ : N√∫mero total de perros vendidos.

$ùëù=0.8$: Probabilidad de que un perro termine en la calle.

$q=1‚àíp=0.2$: Probabilidad de que un perro no termine en la calle.

Media (ùúá) y desviaci√≥n est√°ndar (ùúé) de la distribuci√≥n binomial:

$$\mu = n \cdot p, \quad \sigma = \sqrt{n \cdot p \cdot q}$$

F√≥rmulas clave para la aproximaci√≥n normal:

Utilizamos la distribuci√≥n normal con media ùúá y desviaci√≥n est√°ndar ùúé para aproximar la binomial. Aplicaremos la correcci√≥n por continuidad, sumando o restando 0.5 a los valores enteros.

$$Z = \frac{X - \mu}{\sigma}$$

---

a) Probabilidad de que menos de 30 perros terminen en la calle:

Esto se traduce en: $$ùëÉ(ùëã<30)$$

- Calculamos la media y la desviaci√≥n est√°ndar:

$$\mu = 50 \cdot 0.8 = 40, \quad \sigma = \sqrt{50 \cdot 0.8 \cdot 0.2} \approx 2.828$$

- Aplicamos la correcci√≥n por continuidad $(ùëã=29.5)$:

$$Z = \frac{29.5 - 40}{\sigma} = \frac{29.5 - 40}{2.828} \approx -3.71$$

- Calculamos $ùëÉ(ùëã<30)$ usando la funci√≥n acumulada de la distribuci√≥n normal:

$$P(X<30)=norm.cdf(‚àí3.71)$$

---

b) Probabilidad de que entre 30 y 45 perros terminen en la calle:
Esto se traduce en:

$$P(30‚â§X‚â§45)$$

Aplicamos la correcci√≥n por continuidad:

$$P(30‚â§X‚â§45)=P(X‚â§45.5)‚àíP(X‚â§29.5)$$

Calculamos los valores de ùëç:

Para $X=45.5$:

$$Z = \frac{45.5 - 40}{\sigma} = \frac{45.5 - 40}{2.828} \approx 1.95$$

Para ùëã=29.5 (calculado antes):

$$Z‚âà‚àí3.71$$
Calculamos las probabilidades acumuladas:
$$P(30‚â§X‚â§45)=norm.cdf(1.95)‚àínorm.cdf(‚àí3.71)$$

```python
import scipy.stats as stats
import numpy as np

n = 50  # N√∫mero de perros vendidos
p = 0.8  # Probabilidad de que un perro termine en la calle
q = 1 - p  # Probabilidad de que un perro no termine en la calle

media = n * p
desviacion_estandar = np.sqrt(n * p * q)

probabilidad_menos_30 = stats.norm.cdf(30, loc=media, scale=desviacion_estandar)

probabilidad_entre_30_y_45 = stats.norm.cdf(45, loc=media, scale=desviacion_estandar) - stats.norm.cdf(30, loc=media, scale=desviacion_estandar)

# Imprimir resultados
print(f"Probabilidad de que menos de 30 perros terminen en la calle: {probabilidad_menos_30:.4f}")
print(f"Probabilidad de que entre 30 y 45 perros terminen en la calle: {probabilidad_entre_30_y_45:.4f}")
```

- $n=100$

Media (ùúá) y desviaci√≥n est√°ndar (ùúé) de la distribuci√≥n binomial:

$$\mu = n \cdot p, \quad \sigma = \sqrt{n \cdot p \cdot q}$$

Utilizamos la distribuci√≥n normal con media ùúá y desviaci√≥n est√°ndar ùúé para aproximar la binomial. Aplicaremos la correcci√≥n por continuidad, sumando o restando 0.5 a los valores enteros.

Y aplicamos la correcci√≥n por continuidad, ajustando los l√≠mites $ùëã¬±0.5$.

---

a) Probabilidad de que m√°s de 55 personas vean TV Azteca:

$$P(X>55)=P(X‚â•56)$$

Calculamos la media y la desviaci√≥n est√°ndar:

Usamos la funci√≥n acumulada de la normal est√°ndar:

$$P(X>55)=1‚àíP(Z‚â§3.17)=1‚àínorm.cdf(3.17)$$

b) Probabilidad de que menos de 30 personas sintonicen TV Azteca:
