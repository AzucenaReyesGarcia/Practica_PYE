### 3.1.4 Función acumulada

Es una parte fundamental de la cantidad de estrategias que deben aprenderse para un buen conocimiento de estadística.
 Probabilidad de que X tenga un valor menor o igual que x.

- $P(X ≤ x)$

Nota que:
- $P(X≥x) + P(X<x)=1$
- $P(X>x) + P(X ≤ x)=1$

#### Ejemplo

Supongamos que \(X\) es la variable aleatoria que representa la cantidad de horas que un grupo de estudiantes estudia por semana. La función de distribución acumulada \(F(x)\) se define de la siguiente manera:

- $F(10) = 0.30$ (30% de los estudiantes estudian 10 horas o menos)
- $F(15) = 0.70$ (70% de los estudiantes estudian 15 horas o menos)

Utilizando la función acumulada, podemos calcular:

- **Probabilidad de que un estudiante estudie 10 horas o menos:**
  $P(X \leq 10) = F(10) = 0.30$

- **Probabilidad de que un estudiante estudie más de 10 horas:**
  $P(X > 10) = 1 - P(X \leq 10) = 1 - 0.30 = 0.70$

- **Probabilidad de que un estudiante estudie entre 10 y 15 horas:**
  $P(10 < X \leq 15) = F(15) - F(10) = 0.70 - 0.30 = 0.40$

Esto indica que hay un 40% de probabilidad de que un estudiante estudie entre 10 y 15 horas a la semana.

### 3.2.1 Distribución de probabilidad en forma general

Es aquella que permite establecer toda la gama de resultados probables de ocurrir en un experimento determinado.

Se llama función de densidad a la distribución de probabilidad de una variable aleatoria continua .
Las propiedades que debe cumplir una función de densidad son:
 - $f(x) ≥ 0$, para todos $x ∈ R$

 - ${\displaystyle \int _{-\infty }^{\infty }f_{X}(x)\,dx=1}$

- ${\displaystyle \operatorname {P} [a\leq X\leq b]=\int _{a}^{b}f_{X}(x)\,dx=F(b)-F(a)}$

donde x es el valor de la variable aleatoria y f(x) la probabilidad asociada con el valor x.
Una función de densidad puede representarse de dos formas:
- Gráfica: es una gráfica de línea, cuya área bajo la curva es la probabilidad entre dos valores de la variable aleatoria.
- Como función: cuando la distribución de densidad se define mediante una fórmula.

 d. Varianza

Ahora, calculamos $E(X)$:

$E(X) = \int_{0}^{\sqrt{2}} x \cdot x \, dx = \int_{0}^{\sqrt{2}} x^2 \, dx = \left[ \frac{x^3}{3} \right]_{0}^{\sqrt{2}} = \frac{(\sqrt{2})^3}{3} = \frac{2\sqrt{2}}{3}$

### 3.2.4 Función acumulada



 Es una parte fundamental de la cantidad de estrategias que deben aprenderse para un buen conocimiento de estadística. Probabilidad de qué X tenga un valor menor o igual que x.
 - $F(x) = P(X ≤ x)$
 La probabilidad de que $X$ se sitúe en un intervalo $(a, b]$ (abierto en a y cerrado en $b) es F(b) − F(a)$ si a $≤ b$. Esto es:
 - $P(a < X ≤ b) = Fx (b) - Fx(a)$

 #### Ejemplo

Supongamos que \(X\) es una variable aleatoria que representa la altura de estudiantes en centímetros, y se ha determinado que la función acumulada \(F(x)\) se define de la siguiente manera:


- $F(160) = 0.25\)# (25% de los estudiantes tienen una altura menor o igual a 160 cm)
- \(F(170) = 0.60\) (60% de los estudiantes tienen una altura menor o igual a 170 cm)


Para calcular la probabilidad de que un estudiante tenga una altura entre 160 cm y 170 cm:

$
P(160 < X \leq 170) = F(170) - F(160) = 0.60 - 0.25 = 0.35
$

Esto significa que hay un 35% de probabilidad de que un estudiante tenga una altura entre 160 cm y 170 cm.


### 3.2.5 Cálculos de probabilidad

El cálculo de probabilidades se expresa en porcentaje y responde a la siguiente fórmula:
- Probabilidad = Casos favorables / casos posibles x 100.

#### Componentes

1. **Casos Favorables**: Resultados que cumplen con el evento de interés.
2. **Casos Posibles**: Total de resultados que pueden ocurrir.

#### Ejemplo

Si lanzas un dado y quieres saber la probabilidad de obtener un 4:

- **Casos Favorables**: 1 (el 4).
- **Casos Posibles**: 6 (números del 1 al 6).

Aplicando la fórmula:

$
\text{Probabilidad} = \left( \frac{1}{6} \right) \times 100 \approx 16.67\%
$

Esto indica que hay un 16.67% de probabilidad de obtener un 4 al lanzar el dado.
